```python
import geopandas as gpd
import os

# Load the two GeoDataFrames
points = gpd.read_file('Points.shp')
clli = gpd.read_file('clli_boundary.shp')
MSA = gpd.read_file('tl_2020_us_cbsa.shp')

# Perform a spatial join based on points within polygons


result1 = gpd.sjoin(points, clli, how='left', op='intersects')
result1.to_file('file1.shp', driver='ESRI Shapefile')
# 
file1 = gpd.read_file('file1.shp')

result2 = gpd.sjoin(file1, MSA, how='left', op='intersects')
os.remove('file1.shp')
gdf = result2.drop(columns=['index_right', 'CSAFP','CBSAFP','GEOID','MEMI',  'MTFCC', 'ALAND',  'AWATER' ,'INTPTLAT' ,'INTPTLON',  'lata_nm',   'ocn','ocn_name','c2f_name' ,'LSAD','index_righ' ])
print(result2)

# Print the result

gdf.to_file('Final Output.shp', driver='ESRI Shapefile')
gdf.to_csv('FinalOutput.csv', index=False)
```
